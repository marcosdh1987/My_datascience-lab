```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
#dd2.head()

market_basket = dd1[dd1['PREMIUM_CUSTOMER'] =="Mainstream"].groupby(['DATE','brand_name'])['PROD_QTY']

market_basket = market_basket.sum().unstack().reset_index().fillna(0).set_index('DATE')

market_basket
```

```python
def encode_data(datapoint):
    if datapoint <= 0:
        return 0
    if datapoint >= 1:
        return 1
    
my_basket_sets = market_basket.applymap(encode_data)
my_basket_sets
```

```python
#Generating frequent itemsets

my_frequent_itemsets = apriori(my_basket_sets, min_support=0.15, use_colnames=True)
```

```python
#my_frequent_itemsets.sort_values(by='support',ascending=False).head(80)
my_frequent_itemsets['itemsets'].count()

my_frequent_itemsets['itemsets_count'] = my_frequent_itemsets['itemsets'].apply(len)
dds = my_frequent_itemsets[my_frequent_itemsets['itemsets_count'] == 1]
dds = dds[dds['support'] != 1]
dds.sort_values(by='support',ascending=False).head(30)
#dds = dds.drop(['itemsets_count'],axis=1)
#my_frequent_itemsets.head(60)
```
