```python
import seaborn as sns
import pandas as pd
import earthpy as et

# Handle date time conversions between pandas and matplotlib
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()

# Use white grid plot background from seaborn
sns.set(font_scale=1.5, style="whitegrid")

import matplotlib.pyplot as plt

fig, ax = plt.subplots(8,figsize=(22,15),sharex=True)
ds.plot(y='Static Pressure\n(Average)',color='red',ax=ax[0])
ds.plot(y='Gas Temp\n(Average)',color='orange',ax=ax[1])
ds.plot(y='OilFlowCoriolis\n(Average)',color='green',ax=ax[2])
ds.plot(y='Oil Density\n(Average)',color='gray',ax=ax[3])
ds.plot(y='Oil Temperature\n(Average)',color='cyan',ax=ax[4])
ds.plot(y='Water Flow Corrected\n(Average)',color='blue',ax=ax[5])
ds.plot(y='Water Density\n(Average)',color='blue',ax=ax[6])
ds.plot(x='Date & Time', y='GasFlow_std\n(Average)',color='orange',ax=ax[7])

ax[0].set(title="Pressure [Psig]")
ax[1].set(title="Gas Temp [C]")
ax[2].set(title="Oil Flow [m3/d]")
ax[3].set(title="Oil Denstity [kg/m3]")
ax[4].set(title="Oil Temp [C]")
ax[5].set(title="Water Flow [m3/d]")
ax[6].set(title="Water Density [kg/m3]")
ax[7].set(title="Gas Flow [m3/d]")
plt.show()
```
