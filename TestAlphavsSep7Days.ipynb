```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

data1 = pd.read_csv(r'D:\Google Drive\Deep_Learning\Jupiter_Notebooks\data_science\7Days_SEP_vs_Alpha_1Min.csv')

data2 = data1.copy()
print(data2.iloc[:,3:9])
```

```python
_ = sns.boxplot(x='WELL+CHOKE',y='SEP Gas Flow Rate', data=data1)
_ = plt.xlabel('WELL+CHOKE')
_ = plt.ylabel('SEP Gas Flow Rate')
plt.show()

_ = sns.boxplot(x='WELL+CHOKE',y='ALPHA Gas Flow Rate', data=data1)
_ = plt.xlabel('WELL+CHOKE')
_ = plt.ylabel('ALPHA Gas Flow Rate')
plt.show()
```

```python
_ = sns.boxplot(x='WELL+CHOKE',y='SEP Oil Flow Rate', data=data1)
_ = plt.xlabel('WELL+CHOKE')
_ = plt.ylabel('SEP Oil Flow Rate')
plt.show()

_ = sns.boxplot(x='WELL+CHOKE',y='ALPHA Oil Flow Rate', data=data1)
_ = plt.xlabel('WELL+CHOKE')
_ = plt.ylabel('ALPHA Oil Flow Rate')
plt.show()
```

```python
_ = sns.boxplot(x='WELL+CHOKE',y='SEP Water Flow Rate', data=data1)
_ = plt.xlabel('WELL+CHOKE')
_ = plt.ylabel('SEP Water Flow Rate')
plt.show()

_ = sns.boxplot(x='WELL+CHOKE',y='ALPHA Water Flow Rate', data=data1)
_ = plt.xlabel('WELL+CHOKE')
_ = plt.ylabel('ALPHA Water Flow Rate')
plt.show()
```

```python
mean1 = np.mean(data1['ALPHA Oil Flow Rate'])
std1 = np.std(data1['ALPHA Oil Flow Rate'])
mean2 = np.mean(data1['SEP Oil Flow Rate'])
std2 = np.std(data1['SEP Oil Flow Rate'])
dist1 = np.random.normal(mean1, std1, size=100000)
dist2 = np.random.normal(mean2, std2, size=100000)
plt.hist(dist2, color='red', bins=1000)
plt.hist(dist1, color='blue', bins=1000)
plt.xlabel('Mean')
plt.ylabel('Oil Flow Rate')
plt.show()
```

```python
mean1 = np.mean(data1['ALPHA Gas Flow Rate'])
std1 = np.std(data1['ALPHA Gas Flow Rate'])
mean2 = np.mean(data1['SEP Gas Flow Rate'])
std2 = np.std(data1['SEP Gas Flow Rate'])
dist1 = np.random.normal(mean1, std1, size=100000)
dist2 = np.random.normal(mean2, std2, size=100000)
plt.hist(dist2, color='orange', bins=1000)
plt.hist(dist1, color='green', bins=1000)
plt.xlabel('Mean')
plt.ylabel('Gas Flow Rate')
plt.show()
```

```python
mean1 = np.mean(data1['ALPHA Water Flow Rate'])
std1 = np.std(data1['ALPHA Water Flow Rate'])
mean2 = np.mean(data1['SEP Water Flow Rate'])
std2 = np.std(data1['SEP Water Flow Rate'])
dist1 = np.random.normal(mean1, std1, size=100000)
dist2 = np.random.normal(mean2, std2, size=100000)
plt.hist(dist2, color='black', bins=1000)
plt.hist(dist1, color='blue', bins=1000)
plt.xlabel('Mean')
plt.ylabel('Water Flow Rate')
plt.show()
```
