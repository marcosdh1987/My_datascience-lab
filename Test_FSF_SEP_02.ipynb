```python
ds_24 = ds_data.groupby(ds_data.index // 24)['FSF Line Press [Psig]',
       'FSF Line Temp [C]', 'FSF Gas Rate [sm3/d]', 
       'FSF Oil Rate [sm3/d]', 
       'FSF Water Rate [sm3/d]', 'FSF Liquid [sm3/d]',
       'SEP Gas Rate [sm3/d]', 'SEP Oil Flow [sm3/d]',
       'SEP RAW Oil Flow [m3/d]', 
          'SEP liquid [sm3/d]','SEP Water Flow Rate [sm3/d]'].mean()
```

```python
ds_24['Gas diff (tot)'] = (ds_24['FSF Gas Rate [sm3/d]']-ds_24['SEP Gas Rate [sm3/d]'])/ds_24['SEP Gas Rate [sm3/d]'] 
ds_24['Liq diff (tot)'] = (ds_24['FSF Liquid [sm3/d]']-ds_24['SEP liquid [sm3/d]'])/ds_24['SEP liquid [sm3/d]'] 
ds_24['Oil diff (tot)'] = (ds_24['FSF Oil Rate [sm3/d]']-ds_24['SEP Oil Flow [sm3/d]'])/ds_24['SEP Oil Flow [sm3/d]']
ds_24['Water diff (tot)'] = (ds_24['FSF Water Rate [sm3/d]']-ds_24['SEP Water Flow Rate [sm3/d]'])/ds_24['SEP Water Flow Rate [sm3/d]']
ds_24['Gas diff (abs)'] = (ds_24['FSF Gas Rate [sm3/d]']-ds_24['SEP Gas Rate [sm3/d]']).abs()/ds_24['SEP Gas Rate [sm3/d]']
ds_24['Liq diff (abs)'] = (ds_24['FSF Liquid [sm3/d]']-ds_24['SEP liquid [sm3/d]']).abs()/ds_24['SEP liquid [sm3/d]'] 
ds_24['Oil diff (abs)'] = (ds_24['FSF Oil Rate [sm3/d]']-ds_24['SEP Oil Flow [sm3/d]']).abs()/ds_24['SEP Oil Flow [sm3/d]']
ds_24['Water diff (abs)'] = (ds_24['FSF Water Rate [sm3/d]']-ds_24['SEP Water Flow Rate [sm3/d]']).abs()/ds_24['SEP Water Flow Rate [sm3/d]']

ds_24[['Gas diff (tot)','Liq diff (tot)','Oil diff (tot)','Water diff (tot)','Gas diff (abs)','Liq diff (abs)','Oil diff (abs)','Water diff (abs)','SEP GOR']].describe()
```

```python
ds_d1 = ds_data.copy()

ds_d1['SEP GOR'] = ds_d1['FSF Gas Rate [sm3/d]'] / ds_d1['FSF Oil Rate [sm3/d]']  


x1 = ds_d1.query('WELL == "Lca-3083"')
x2 = ds_d1.query('WELL == "Lca-3084"')
x3 = ds_d1.query('WELL == "Lca-3085"')


import seaborn as sns
import matplotlib.pyplot as plt
sns.kdeplot(x1['SEP GOR'], shade=True, label='Lca-3083')
sns.kdeplot(x2['SEP GOR'], shade=True, label='Lca-3084')
sns.kdeplot(x3['SEP GOR'], shade=True, label='Lca-3085')
plt.title("'SEP GOR'")
```
