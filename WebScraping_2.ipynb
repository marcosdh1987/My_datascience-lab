```python
from urllib.request import Request, urlopen
from bs4 import BeautifulSoup
#url = 'https://www.clarin.com/astrologia'
#url = 'https://www.datamind.cl/'
url = 'https://resultados.as.com/resultados/futbol/primera/clasificacion'

req = Request(url , headers={'User-Agent': 'Mozilla/5.0'})
#req = Request(url)

webpage = urlopen(req).read()
soup = BeautifulSoup(webpage, "html.parser")
title = soup.find("title")
print(title)
```

```python
containers = soup.findAll("p")
for container in containers:
    print(container)
```

```python
# Print out the text
text = soup.get_text()
#print(soup.text)
```

```python
all_links = soup.find_all("a")
for link in all_links:
    print(link.get("href"))
```

```python
# Print the first 10 rows for sanity check
rows = soup.find_all("tr") #options: tr,meta
print(rows[:5])
```

```python
#row_td = []

for row in rows:
     row_td = row.find_all("td")

print(row_td)
type(row_td)
```

```python
str_cells = str(row_td)
cleantext = BeautifulSoup(str_cells, "lxml").get_text()
print(cleantext)
```

```python
import re

list_rows = []
for row in rows:
    cells = row.find_all('td')
    str_cells = str(cells)
    clean = re.compile('<.*?>')
    clean2 = (re.sub(clean, '',str_cells))
    list_rows.append(clean2)
print(clean2)
type(clean2)
```

```python
import pandas as pd
df = pd.DataFrame(list_rows)
df.head(10)
```

```python
#equipos

eq = soup.find_all('span',class_='nombre-equipo')
pt = soup.find_all('td',class_='destacado')
#print(eq)

equipos = list()
puntos = list()

for i in eq:
    equipos.append(i.text)
for i in pt:
    puntos.append(i.text)   

df = pd.DataFrame(equipos,columns=['Equipo'])
df['Puntos'] = pd.DataFrame(puntos)

print(df.Equipo.unique())
print(df.head())
```
